```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pywt
img=cv2.imread('sun_face/sunny3.jpeg')
plt.subplot(121)
plt.imshow(img)
img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
img=np.float32(img)
img/=255
#plt.figure(figsize=(20,20))
coef=pywt.wavedec2(img,'haar',level=5)
coef_list=list(coef)
coef_list[0]*=0

jpg=pywt.waverec2(coef_list,'haar')
jpg*=255
jpg=np.uint8(jpg)
plt.subplot(122)
plt.imshow(jpg,cmap='gray')

```
